```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
xrp = pd.read_csv(r'C:\Users\SHARVARI AVHAD\Downloads\XRP-INR.csv')
xrp
```

```python
pd.isnull(xrp['Volume']) #checking for null values
```

```python
xrp.dropna() #dropping the rows with null values
```

```python
# Extracting Month and Year from 'Date' Column

xrp['Month'] = pd.DatetimeIndex(xrp['Date']).month
xrp['Year'] = pd.DatetimeIndex(xrp['Date']).year
xrp['Month-Year'] = pd.DatetimeIndex(xrp['Date']).strftime('%m-%Y')

# Dropping the extra unrequired column

xrp.drop("Year-Month", axis="columns", inplace=True)

xrp.head()
```

```python
# Line Chart Of Closing Prie 

fig = px.line(xrp, x="Date", y="Close", labels={"Close":"Closing Price"}, title="Closing Price Trend")

fig.show()
```

#### This Line graph shows that the Highest value of Ripple(XRP) was achieved in mid of April. The price then was approx 138 INR. This graph also amazes us as the the price of Ripple(XRP) Coin went drastically upwards from July 2020 to April 2021.

```python
# 

fig = px.bar(xrp, x="Date", y="Volume", title="Volume of Coins being Sold")

fig.show()
```

```python
# Comparison of Open and Close Values for a partcular Month 

fig = px.bar(xrp,
            x = "Month-Year",
            y = ["Open", "Close"],
            barmode = "group",
            title = "Open and Close Values of XRP"
            )
fig.show()
```

#### This graph shows us that the months in which the Open Value is Higher than the Close Value, there is certain amounth of loss faced by the Investors amd vice versa.
